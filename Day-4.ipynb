# Exploratory Data Analysis (EDA)


### It is the process of analyzing and summarizing data sets to understand their main characteristics, uncover patterns, detect anomalies, and identify relationships between variables. EDA helps to gain insights and develop an understanding of the data before diving into more complex analysis or modeling tasks.

## Here are some key aspects and techniques involved in EDA:

* Data Summary: Understanding the structure, dimensions, and basic statistical properties of the data using summary statistics like mean, median, standard deviation, etc.

* Data Visualization: Creating visual representations of the data through various plots, charts, and graphs to identify patterns, trends, outliers, and relationships. Examples include histograms, box plots, scatter plots, bar charts, etc.

* Handling Missing Values: Identifying missing values in the data and deciding on appropriate strategies for handling them, such as imputation or deletion.

* Data Cleaning: Cleaning the data by removing duplicates, correcting inconsistencies, and handling outliers to ensure data quality.

* Feature Selection: Identifying relevant features or variables that have the most impact on the target variable and eliminating irrelevant or redundant features.

* Feature Engineering: Creating new features from existing ones to enhance the predictive power of the data. This may involve transformations, aggregations, or creating interaction terms.

* Correlation Analysis: Examining the relationships between variables to determine how they are related and whether there are any strong correlations or dependencies.

* Hypothesis Testing: Conducting statistical tests to validate or reject hypotheses about the data and assess the significance of relationships or differences.

### The goal of EDA is to gain insights, generate hypotheses, and inform subsequent analysis or modeling tasks. It helps to ensure data quality, discover patterns, and make informed decisions throughout the data analysis process.

# Graph Plotting with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib
print(matplotlib.__version__)
```

```python
import numpy as np
```

```python
x = np.array([0,10])
y = np.array([0,300])
```

```python
plt.plot(x,y)
plt.show()
```

```python
#labels
plt.xlabel('feq')
plt.ylabel('count')
plt.plot(x,y)
plt.show()
```

```python
#labels | point
plt.xlabel('feq')
plt.ylabel('count')
plt.plot(x, y, 'o')
plt.show()
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([2,4,6,8,10,12,14,16])
y.shape
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
y = np.array([1,47,20,44,1,90])
plt.plot(y, marker = 'o')
plt.show()
```

```python
#fmt and ms(marker size) and mec(points border color)
plt.plot(y, 'o:b', ms = 15, mec='r')
```

```python
#bar graph 
plt.bar([0.25,0.50,0.80,1.0],[20,30,40,80],label = 'Population', color='r', width=0.1)
```

```python
#histogram
age = [12,14,16,20,42,82,72]
bins = [0,10,40,100]
```

```python
plt.hist(age,bins=bins, histtype='bar', rwidth=0.6)
plt.show()
```

```python
#sctterplot
w = np.array([1,2,3,4,5,6,7,8])
s = np.array([8,7,6,5,4,3,2,1])
```

```python
plt.scatter(w,s)
plt.show()
```

```python
a = np.array([12,32,34,54,67,58,45,70])
```

```python
plt.plot([],[],color = 'g', label = w, linewidth=5)
plt.plot([],[],color = 'r', label = s, linewidth=5)
# plt.plot([],[],color = 'g', lable = w, linewidth=5)
plt.stackplot(w,s,a, color = ['c','r'])
```

```python
sales = [12,13,29,30,16]
dept = ['it','ce','hr','cto','tech']
color = ['c','g','b','m','r']
```

```python
plt.pie(sales,labels=dept,colors=color)
plt.show()
```
